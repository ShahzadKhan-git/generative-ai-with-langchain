```python
import os

from langchain_groq import ChatGroq
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

model = ChatGroq(model="qwen/qwen3-32b")
```

```python
from pydantic import BaseModel,Field

class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="Year on which movie is released")
    director:str=Field(description="Director of the movie")
    rating:float=Field(description="Rating of this movie out of 10")
```

```python
model_with_structure=model.with_structured_output(Movie)
model_with_structure
```

```python
model.invoke("Provide the details of movie Interstellar")
```

```python
model_with_structure.invoke("Give details of movie Tenet")
```
