```python
from langchain_groq import ChatGroq
import os

model = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model="llama-3.3-70b-versatile"
)
```

```python
response = model.invoke("what you think about the war happening in Iran with Israel?Who is right in this and what is US doing in this war ?")
print(response.content)
```

```python
response = model.invoke(
    "What do you think about the war happening between Iran and Israel? Who is right and what is the US doing?"
)
print(response.content)
```

```python
model.invoke("Write an essay about mountains in 500 words")
```

```python
for chunk in model.stream("Write an essay about mountains in 500 words"):
    print(chunk.text, end="/", flush=True)
```

```python
responses = model.batch([
    "What is the capital of France?",
    "How airoplanes fly?",
    "What is the meaning of life?"
]
)

for response in responses:
    print(response.content)
```
