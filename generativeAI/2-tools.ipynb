```python
import os 
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model=init_chat_model("groq:llama-3.3-70b-versatile")

response=model.invoke("Write an essay on AI")
print(response.content)
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str)->str:
    """Get the weather at a location"""
    return f"It's sunny in {location}"


model_with_tools=model.bind_tools([get_weather])
```

```python
response=model_with_tools.invoke("What's the weather in Pune?")
print(response)
```

```python
response.tool_calls
```

```python
messages = [{"role" : "user", "content" : "What's the weather in Pune?"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tools_result = get_weather.invoke(tool_call)
    messages.append(tools_result)
    
final_response = model_with_tools.invoke(messages)
print(final_response.content)    
```
